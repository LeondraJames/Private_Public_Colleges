```
#Load data
df = sqlContext.sql("SELECT * FROM college_dataset")
```

```
df.printSchema() #Will be predicting whether a school is private or not (binary classification)
```

```
df.show()
```

```
from pyspark.ml.feature import VectorAssembler
```

```
#Create vector of all predictors
df.columns
```

```
assembler = VectorAssembler(inputCols=[ 'Apps',
 'Accept',
 'Enroll',
 'Top10perc',
 'Top25perc',
 'F_Undergrad',
 'P_Undergrad',
 'Outstate',
 'Room_Board',
 'Books',
 'Personal',
 'PhD',
 'Terminal',
 'S_F_Ratio',
 'perc_alumni',
 'Expend',
 'Grad_Rate'], outputCol='features')
```

```
output = assembler.transform(df)
```

```
#Index the response variable as binary (double)
from pyspark.ml.feature import StringIndexer
```

```
indexer = StringIndexer(inputCol= 'Private', outputCol='PrivateIndex')
```

```
output_fixed = indexer.fit(output).transform(output)
```

```
output_fixed.printSchema()
```

```
#Select (double) response variable & features vector
final_data = output_fixed.select('features','PrivateIndex')
```

```
#Partition
train_data, test_data = final_data.randomSplit([0.7,0.3])
```

```
#Since this is a classifcation problem, we will apply multiple tree-based classification models
from pyspark.ml.classification import DecisionTreeClassifier, GBTClassifier, RandomForestClassifier
```

```
#Create pipeline to run multiple models simultaneously
from pyspark.ml import Pipeline
dtc = DecisionTreeClassifier(labelCol = 'PrivateIndex', featuresCol = 'features')
rfc = RandomForestClassifier(labelCol = 'PrivateIndex', featuresCol = 'features', numTrees=150)
gbt = GBTClassifier(labelCol='PrivateIndex', featuresCol = 'features')
```

```
dtc_model = dtc.fit(train_data)
rfc_model = rfc.fit(train_data)
gbt_model = gbt.fit(train_data)
```

```
dtc_preds = dtc_model.transform(test_data)
rfc_preds = rfc_model.transform(test_data)
gbt_preds = rfc_model.transform(test_data)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
my_binary_eval = BinaryClassificationEvaluator(labelCol = 'PrivateIndex')
```

```
#Decision Tree Performance
print('DTC')
print(my_binary_eval.evaluate(dtc_preds))
```

```
#Random Forest Performance
print('RFC')
print(my_binary_eval.evaluate(rfc_preds))
```

```
#Gradient Boosted Trees Performance - First preview resulting dataframe
gbt_preds.printSchema()
```

```
gbt_preds.show()
```

```
acc_eval = MulticlassClassificationEvaluator(labelCol='PrivateIndex', metricName='accuracy')
```

```
rfc_acc = acc_eval.evaluate(rfc_preds)
```
